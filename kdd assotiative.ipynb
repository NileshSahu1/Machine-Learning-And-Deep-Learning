```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('Market_Basket_Optimisation.csv',header=None)
data.head(10)
```

```python
data.shape
```

```python
data.values.tolist()   #dont work need in array format
```

```python
transactions = []   # array
for i in range(0, 7501):
    transactions.append([str(data.values[i,j]) for j in range(0, 20)])
```

```python
from apyori import apriori
```

```python
pip install apyori 
```

```python
rules=apriori(transactions, min_support=0.003, min_confidence=0.2, 
              min_lift=3,min_length=2)
```

```python
rules
```

```python
results=list(rules)
results
```

```python
results=pd.DataFrame(results)
results.head()
```

```python
list_item_base=[]
list_item_add=[]
list_confidence=[]
list_lift=[]

for i in range(len(results)):
    list_item_base.append(results.loc[i,"ordered_statistics"][0][0])
    list_item_add.append(results.loc[i,"ordered_statistics"][0][1])
    list_confidence.append(results.loc[i,"ordered_statistics"][0][2])
    list_lift.append(results.loc[i,"ordered_statistics"][0][3])
   
results["item_base"]= list_item_base
results["item_add"]= list_item_add
results["confidence"] = list_confidence
results["lift"] = list_lift

results
```
