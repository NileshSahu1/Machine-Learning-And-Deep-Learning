```python
# start with one review:
text= df.description[0]
#ncreate an generate a wordcloud image:
wordcloud = WordCloud().generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#lower max_font_size,change the maximum number of word
wordcloud = WordCloud(max_font_size=50,max_words=100,background_color='white').generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# save the image in the img folder
wordcloud.to_file("first_review.png")
```

```python
wordcloud = WordCloud(max_font_size=50,max_words=100,background_color='white').generate(text)
plt.imshow(wordcloud,interpolation='nearest')
plt.axis('off')
plt.show()
```

```python
stopwords =set(STOPWORDS)
stopwords.update(["drink","now","wine","flavor","flavors"])
wordcloud=WordCloud(stopwords=stopwords,background_color='white').generate(text)
plt.imshow(wordcloud,interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from PIL import Image
wine_mask =np.array(Image.open("wine_mask.png"))
#use pillow to open the image 
wine_mask
```

```python
def transform_format(val):
    if val == 0:
        return 255
    else:
        return val
```

```python
transformed_wine_mask = np.ndarray((wine_mask.shape[0],wine_mask.shape[1]),np.int32)
for i in range(len(wine_mask)):
    transformed_wine_mask[i] = list(map(transform_format, wine_mask[i]))
transformed_wine_mask    
```

```python
# create a wordcloud image
wc =WordCloud(background_color="white",max_words=1000,mask=transformed_wine_mask,stopwords=stopwords,
             contour_width=3,contour_color='firebrick')
wc.generate(text)
wc.to_file("wine.png")
plt.figure(figsize=[20,10])
plt.imshow(wc,interpolation='bilinear')
plt.axis('off')
plt.show()
```
